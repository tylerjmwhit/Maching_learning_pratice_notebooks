```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from datetime import datetime
```

```python
ticker = yf.Ticker("NVDA")
df = ticker.history(period='max')
```

```python
df
```

```python
df.loc[:,"average"] = (df['Open'] + df['High'] + df['Low'] + df['Close'])/4
df.loc[:,'tommorow'] = df['average'].shift(-1)
df.loc[:,'yesterday'] = df['average'].shift(1)
df.loc[:,'roll 3'] = df['average'].rolling(window=3).mean()
df.loc[:,'Dates'] = df.index
```

```python
df =df.dropna()
df.index = range(len(df))
```

```python
df
```

```python
x_train = df[['average','roll 3','yesterday']].values
y_train = df[['tommorow']].values
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)
```

```python
from sklearn.model_selection import KFold
lr = LinearRegression()
kf = KFold(n_splits=5, shuffle=True)

scores = []
for train_index, test_index in kf.split(x_train):
  x_split, x_test = x_train[train_index] , x_train[test_index]
  y_split, y_test = y_train[train_index] , y_train[test_index]
  lr.fit(x_split,y_split)
  scores.append(lr.score(x_test,y_test))
print(scores)
print(np.mean(scores))
```
