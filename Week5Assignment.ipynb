```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

TASK 1

```python
def my_fun(x,y): # defintino of z
  return np.sin(x) + np.cos(y)

def grad_x(x,y):
  return np.cos(x)

def grad_y(x,y):
  return -np.sin(y)
```

```python
 #generating all my points
 x = np.linspace(-5,5, num = 50)
 y = np.linspace(-5,5, num = 50)
 X,Y = np.meshgrid(x,y)
 Z = my_fun(X,Y)
```

```python
#2d landscape of the function
plt.imshow(Z)
```

```python
x0, y0 = np.random.random(2)*5
learning_rate = .8
xxxxs = [x0]
yyyys = [y0]
errors = [my_fun(x0,y0)]
x0,y0
```

```python
for i in range(10000):
  x_grad = grad_x(x0,y0)
  y_grad = grad_y(x0,y0)
  x0 += -x_grad * learning_rate
  y0 += -y_grad * learning_rate
  xxxxs.append(x0)
  yyyys.append(y0)
  errors.append(my_fun(x0,y0))
plt.plot(xxxxs, yyyys)
```

```python
fig = plt.figure()
threeD = plt.axes(projection='3d')
threeD.plot(xxxxs,yyyys,errors,'ro',alpha = 0.5)
threeD.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
threeD.set_title('surface');
threeD.view_init(60, 35)
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
setosa = iris.data[:50]
def f(a1,a2,a3,a4):
    return ((setosa[:,0]-a1)**2 + (setosa[:,1]-a2)**2 + (setosa[:,2]-a3)**2 + (setosa[:,3]-a4)**2).mean()

def grad_a1(a1,a2,a3,a4):
    return np.mean(2 * (setosa[:,0]-a1))

def grad_a2(a1,a2,a3,a4):
    return np.mean(2* (setosa[:,1]-a2))

def grad_a3(a1,a2,a3,a4):
    return np.mean(2*(setosa[:,2]-a3))

def grad_a4(a1,a2,a3,a4):
    return np.mean(2*(setosa[:,3]-a4))


a1 = 0
a2 = 0
a3 = 0
a4 = 0
```

```python
a1s = [a1]
a2s = [a2]
a3s = [a3]
a4s = [a4]
errors = [f(a1,a2,a3,a4)]
```

```python
def grad_descent(a1,a2,a3,a4,a1s,a2s,a3s,a4s):
    a1_grad = grad_a1(a1,a2,a3,a4)
    a2_grad = grad_a2(a1,a2,a3,a4)
    a3_grad = grad_a3(a1,a2,a3,a4)
    a4_grad = grad_a4(a1,a2,a3,a4)
    # print(a1_grad,a2_grad,a3_grad,a4_grad)
    a1 += a1_grad * learning_rate
    a2 += a2_grad * learning_rate
    a3 += a3_grad * learning_rate
    a4 += a4_grad * learning_rate
    a1s.append(a1)
    a2s.append(a2)
    a3s.append(a3)
    a4s.append(a4)
    errors.append(f(a1,a2,a3,a4))
    return a1, a2 ,a3 , a4
```

```python
learning_rate = .1
a1 = 0
a2 = 0
a3 = 0
a4 = 0
a1s = [a1]
a2s = [a2]
a3s = [a3]
a4s = [a4]
errors = [f(a1,a2,a3,a4)]
for i in range(10):
  a1,a2,a3,a4 = grad_descent(a1,a2,a3,a4,a1s,a2s,a3s,a4s)
print(iris.data[:50].mean(axis=0) - [a1,a2,a3,a4])
```
